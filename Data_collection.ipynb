```python
def scrape_images(url, prefix):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    images = soup.find_all('img')

    for index, item in enumerate(images):
        link = item['src']
        if not link.startswith(('http:', 'https:')):
            # URL lacks protocol, so add 'http:' in front
            link = 'http:' + link

        try:
            # Get the image content
            im = requests.get(link)
            if im.status_code == 200:
                # Generate a unique file name using the prefix and index
                file_name = f"{prefix}_{index}.jpg"
                with open(file_name, 'wb') as f:
                    f.write(im.content)
                print(f'Saved image: {file_name}')
            else:
                print(f"Failed to download image {link}. Status code: {im.status_code}")

        except Exception as e:
            print(f"An error occurred while processing {link}: {e}")

```

```python
# making a graph to see the distribution of image sizes of all the images in the dataset




root_directory = '/content/gdrive/MyDrive/data_scraping'

# Initialize lists to store image dimensions
widths = []
heights = []

# Traverse the directory tree and collect image dimensions
for root, _, files in os.walk(root_directory):
    for filename in files:
        if filename.endswith('.jpg') or filename.endswith('.png'):  # Filter for image file extensions
            image_path = os.path.join(root, filename)
            with Image.open(image_path) as img:
                width, height = img.size
                widths.append(width)
                heights.append(height)

```

```python
# Create a scatter plot to visualize the size distribution
plt.figure(figsize=(5, 4))
plt.scatter(widths, heights, marker='o', alpha=0.5)
plt.title('Image Size Distribution')
plt.xlabel('Width (pixels)')
plt.ylabel('Height (pixels)')
plt.grid(True)
plt.show()
```
